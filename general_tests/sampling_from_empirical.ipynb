```python
import pandas as pd
import numpy as np
from pathlib import Path
import plotly.graph_objects as go

df = pd.read_csv(Path.cwd().parent / "data" / "OHCL" / "snp500.csv", index_col=1)
returns = df["Close"].pct_change().values[1:]
fig = go.Figure()
fig.add_trace(go.Histogram(x=returns))
fig.show()
```

```python
cum_returns = sorted(returns)
n = len(cum_returns)
cum_prob = np.arange(1, n+1) / n
fig = go.Figure()
fig.add_trace(go.Scatter(x=cum_returns, y=cum_prob))
fig.show()
```

```python
def sample(sorted_returns):
    n = len(sorted_returns)
    u = np.random.uniform()
    for i,val in enumerate(sorted_returns):
        if i >= n*u:
            return val

def samples(sorted_returns, amount):
    return [sample(sorted_returns) for _ in range(amount)]

N = 10000
s = samples(cum_returns, N)
print(s)

fig = go.Figure()
fig.add_trace(go.Histogram(x=s))
fig.show()

# looks pretty good
```
